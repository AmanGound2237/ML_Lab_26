```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
import seaborn as sns
titanic_data = sns.load_dataset('titanic')
```

```python
missing_values = titanic_data.isnull().sum()
print(missing_values)
```

```python
new_titanic_df = titanic_data.drop(columns = ['deck'])
```

```python
new_titanic_df['age'].fillna(new_titanic_df['age'].median(),inplace=True)
```

```python
missing_values = new_titanic_df.isnull().sum()
print(missing_values)
```

```python
data = new_titanic_df
```

```python
data['embark_town'].dtype
```

```python
data['embark_town'].unique()
```

```python

data['embark_town'].fillna(data['embark_town'].mode()[0] , inplace=True)
```

```python
data.isnull().sum()
```

```python
data['embarked'].dtype
```

```python
data['embarked'].unique()
```

```python

data['embarked'].fillna(data['embarked'].mode()[0] , inplace=True)
```

```python
data.isnull().sum()
```

```python
le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
data['embarked'] = le.fit_transform(data['embarked'])
```

```python
data.head()
```

```python
data.info()
```

```python
data = data[['pclass', 'sex' , 'age' , 'fare' , 'embarked' , 'survived' ]]
X = data[['pclass', 'sex' , 'age' , 'fare' , 'embarked' ]]
Y = data['survived']
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.3 , random_state = 4)
```

```python
model = LogisticRegression(max_iter = 1000)
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test,Y_pred)
print("Accuracy:",accuracy)
```

```python
new_passenger = pd.DataFrame({
    'pclass' : [3],
    'sex' : ['male'],
    'age' : [28],
    'fare' : [7.25],
    'embarked' : ['S']
})
```

```python
new_passenger_encoded = pd.get_dummies(new_passenger)
new_passenger_encoded = new_passenger_encoded.reindex(columns=X.columns , fill_value =0)
```

```python
prediction = model.predict(new_passenger_encoded)
print("Survived" if prediction[0] == 1 else "Did not survive")
```

```python
import pandas as pd
import numpy as np
titanic_df = sns.load_dataset('titanic')

mean_age = np.mean(titanic_df['age'])
std_dev_age = np.std(titanic_df['age'])
Z_scores_age = (titanic_df['age'] - mean_age) / std_dev_age
outliers_age = titanic_df['age'][np.abs(Z_scores_age) > 3]
print("Outliers in 'Age' using Z-score: \n", outliers_age)


mean_fare = np.mean(titanic_df['fare'])
std_dev_fare = np.std(titanic_df['fare'])
Z_scores_fare = (titanic_df['fare'] - mean_fare) / std_dev_fare
outliers_fare = titanic_df['fare'][np.abs(Z_scores_fare) > 3]
print("\nOutliers in 'Fare' using Z-score: \n", outliers_fare)
```
